```python
import subprocess

def search_dix_file(filename, query):
    # Construct the lt-expand command with the filename and pipe it to grep
    command = f"lt-expand {filename} | grep -E '{query}'"
    
    # Execute the command and capture the output
    output = subprocess.check_output(command, shell=True, text=True)
    
    # Return the output
    return output
```

```python
filename = "tel_apertium_v1.1.dix"
query = "^[^:]+:pilla<"

results = search_dix_file(filename, query)

print(results)
```

```python
words = [line.split(":")[0] for line in results.split("\n")]
print(words)
```
