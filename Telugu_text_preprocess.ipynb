```python
parent_folder_path = './texts/'
```

```python
import os
import pandas as pd

# define parent folder path

# define list to store results
results = []

# loop through each folder in the parent folder
for folder_name in os.listdir(parent_folder_path):
    folder_path = os.path.join(parent_folder_path, folder_name)
    
    # check if folder contains text files
    if not any(file.endswith('.txt') for file in os.listdir(folder_path)):
        continue
    
    # loop through each text file in the folder
    for file_name in os.listdir(folder_path):
        if not file_name.endswith('.txt'):
            continue
        
        # read text file as a list of strings and remove empty lines
        with open(os.path.join(folder_path, file_name), 'r') as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]

        # create dictionary with line numbers, text, and folder name
        data_dict = {'line_number': [i+1 for i in range(len(lines))],
                     'text': lines,
                     'folder_name': [folder_name] * len(lines)}
        
        # convert dictionary to pandas dataframe and append to results list
        df = pd.DataFrame.from_dict(data_dict)
        results.append(df)

# concatenate all dataframes in the results list into a single dataframe
df = pd.concat(results, ignore_index=True)

# display dataframe
print(df)
```

```python
import os
import pandas as pd

# define parent folder path

# define list to store results
results = []

# loop through each folder in the parent folder
for folder_name in os.listdir(parent_folder_path):
    folder_path = os.path.join(parent_folder_path, folder_name)
    
    # check if folder contains text files
    if not any(file.endswith('.txt') for file in os.listdir(folder_path)):
        continue
    
    # loop through each text file in the folder
    for file_name in os.listdir(folder_path):
        if not file_name.endswith('.txt'):
            continue
        
        # read text file as a list of strings and remove empty lines
        with open(os.path.join(folder_path, file_name), 'r') as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]

        # create dictionary with line numbers, text, and folder name
        data_dict = {'line_number': [i+1 if line.strip() else "" for i, line in enumerate(lines)],
                     'text': [line for line in lines if line.strip()],
                     'folder_name': [folder_name] * len(lines)}
        
        # convert dictionary to pandas dataframe and append to results list
        df = pd.DataFrame.from_dict(data_dict)
        results.append(df)

# concatenate all dataframes in the results list into a single dataframe
df = pd.concat(results, ignore_index=True)

# display dataframe
print(df)
```
